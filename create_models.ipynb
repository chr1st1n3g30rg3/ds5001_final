```python
def gather_docs(CORPUS, ohco_level, term_col='term_str'):
    OHCO = CORPUS.index.names
    CORPUS[term_col] = CORPUS[term_col].astype('str')
    DOC = CORPUS.groupby(OHCO[:ohco_level])[term_col].apply(lambda x:' '.join(x)).to_frame('doc_str')
    return DOC
```

```python
def compute_pca(X, k, norm_docs, center_by_mean, center_by_variance):
    """
    Compute PCA from a given document-term count matrix.

    Parameters:
        X (pd.DataFrame): The input document-term matrix.
        k (int): The number of principal components to generate.
        norm_docs (bool): If True, normalize documents using L2 norm.
        center_by_mean (bool): If True, mean-center the matrix.
        center_by_variance (bool): If True, standardize by variance.

    Returns:
        LOADINGS (pd.DataFrame): Term-component matrix.
        DCM (pd.DataFrame): Document-component matrix.
        COMP (pd.DataFrame): Component information table.
    """
    # Normalize documents using L2 norm if specified
    if norm_docs:
        X = (X.T / norm(X, 2, axis=1)).T

    # Drop NaN rows and track loss
    initial_rows = len(X)
    X = X.dropna()
    bag_loss = initial_rows - len(X)

    # Mean-center the matrix if specified
    if center_by_mean:
        X = X - X.mean()

    # Standardize by variance if specified
    if center_by_variance:
        X = X / X.std(ddof=0)

    # Compute covariance matrix
    COV = X.cov()

    # Compute eigenvalues and eigenvectors
    eig_vals, eig_vecs = eigh(COV)

    # Store eigenvectors and eigenvalues
    EIG_VEC = pd.DataFrame(eig_vecs, index=COV.index, columns=COV.index)
    EIG_VAL = pd.DataFrame(eig_vals, index=COV.index, columns=['eig_val'])
    EIG_VAL.index.name = 'term_str'

    # Compute explained variance
    EIG_PAIRS = EIG_VAL.join(EIG_VEC.T)
    EIG_PAIRS['exp_var'] = np.round((EIG_PAIRS.eig_val / EIG_PAIRS.eig_val.sum()) * 100, 2)

    # Select top k principal components
    COMP = EIG_PAIRS.sort_values('exp_var', ascending=False).head(k).reset_index(drop=True)
    COMP.index.name = 'comp_id'
    COMP.index = ["PC{}".format(i + 1) for i in range(k)]
    COMP.index.name = 'pc_id'

    # Compute loadings (term-component matrix)
    LOADINGS = COMP[COV.index].T
    LOADINGS.index.name = 'term_str'

    # Compute document-component matrix
    DCM = X.dot(COMP[COV.index].T)

    return LOADINGS, DCM, COMP
```

```python
def vis_pcs(M, a, b, label, hover_name=None, symbol=None, size=None):
    M = M.reset_index()  # Ensure index is a column
    if hover_name is None:
        hover_name = M.index.name if M.index.name else "index"  # Default to index if not specified
    return px.scatter(M, f"PC{a}", f"PC{b}", color=label, hover_name=hover_name, 
                     symbol=symbol, size=size,
                     marginal_x='box', height=800)
```

```python
def vis_loadings(loadings_df, a=0, b=1, hover_name='term_str'):
    X = loadings_df.join(VOCAB_REDUCED)  # Using the passed loadings dataframe
    return px.scatter(X.reset_index(), f"PC{a}", f"PC{b}", 
                      text=hover_name, size='n', color='max_pos', 
                      marginal_x='box', height=800)
```
